```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randn(100000)
```

```python
hx, hy, _ = plt.hist(data, bins=50, color="red")
plt.ylim(0.0,max(hx)+0.05)
plt.title('Generate random numbers \n from a normal distribution with python')
plt.grid()
plt.show()
```
